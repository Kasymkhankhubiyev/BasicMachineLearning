**Среднее значение и среднее отклонение возраста тех, кто получает среднюю зарплату более 50к и тех, кто получает менее 50к**

```python
mean_more_than_50 = data[(data['salary'] == ' >50K')]['age'].mean()

mean_less_than_50 = data[(data['salary'] == ' <=50K')]['age'].mean()

std_more_than_50 = data[(data['salary'] == ' >50K')]['age'].std()

std_less_than_50 = data[(data['salary'] == ' <=50K')]['age'].std()

print(f'более 50К: среднее значение: {mean_more_than_50}, квадратичное отклонение: {std_more_than_50}')
print(f'менее 50К: среднее значение: {mean_less_than_50}, квадратичное отклонение: {std_less_than_50}')
```

**Среди кого больше доля зарабатывающих более 50к женатых или холостых мужчин?**

```python
df = data[data['salary'] == ' >50K']

dt = df.groupby(['marital-status'])['age'].describe()

dt
```

```python
total_married = dt.loc[' Married-AF-spouse']['count'] + dt.loc[' Married-civ-spouse']['count'] + dt.loc[' Married-spouse-absent']['count']

total_free = dt.loc[' Divorced']['count'] + dt.loc[' Separated']['count'] + dt.loc[' Never-married']['count'] + dt.loc[' Widowed']['count']

total = total_married + total_free

print(f' Доля женатых = {total_married/total}')
print(f' Доля холостых =  {total_free/total}')
```
